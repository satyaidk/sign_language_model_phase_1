```python
import json
import cv2
import numpy as np
import tensorflow as tf
```

```python
model = tf.keras.models.load_model("../models/classify_model.keras")
```

```python
with open("../models/class_indices.json", "r") as f:
    class_indices = json.load(f)

# Reverse mapping: index â†’ label
index_to_label = {v: k for k, v in class_indices.items()}

print(index_to_label)
```

```python
img_path = r"C:\Users\nikad\OneDrive\Desktop\as_language_model\testing\london.jpeg"

img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.resize(img, (299, 299))
img = img / 255.0
img = np.expand_dims(img, axis=0)
```

```python
pred = model.predict(img)
class_index = np.argmax(pred)
alphabet = index_to_label[class_index]

confidence = np.max(pred)

print(f"Predicted Alphabet: {alphabet}")
print(f"Confidence: {confidence:.2f}")
```

```python
import cv2
import json
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.inception_resnet_v2 import preprocess_input
```

```python
model = tf.keras.models.load_model("../models/classify_model.keras")

with open("../models/class_labels.json") as f:
    class_labels = json.load(f)

print("Model and labels loaded")
```

```python
img = cv2.imread(
    r"C:\Users\nikad\OneDrive\Desktop\as_language_model\testing\images.jpg"
)

img = cv2.resize(img, (299, 299))
img = np.expand_dims(img, axis=0)
img = preprocess_input(img)

pred = model.predict(img)
pred_idx = str(np.argmax(pred))

print("Predicted Alphabet:", class_labels[pred_idx])
print("Confidence:", round(float(np.max(pred)) * 100, 2), "%")
```

```python
img_show = cv2.imread(
    r"C:\Users\nikad\OneDrive\Desktop\as_language_model\testing\images.jpg"
)

cv2.putText(
    img_show,
    f"Sign: {class_labels[pred_idx]}",
    (20, 40),
    cv2.FONT_HERSHEY_SIMPLEX,
    1.2,
    (0, 255, 0),
    3
)

cv2.imshow("Prediction", img_show)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
